### Query in/out flow info.

```python
import pandas as pd
import datetime
import numpy as np

query = '''
SELECT * FROM in_out
ORDER BY id, time;
'''

flow = pd.read_sql_query(query, con=engine)
# flow['time'] = flow['time'] - datetime.datetime(2017, 7, 1, 0, 0, 0)
# flow['time'] = flow['time'] / np.timedelta64(1, 's')
# flow['time'] = flow[['time']].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
flow
```

### Select time and out-flow

```python
ls = {}
for index, row in flow.iterrows():
    if row['id'] not in ls:
        ls[row['id']] = pd.DataFrame(columns=['time', 'out'])
    ls[row['id']] = ls[row['id']].append(pd.DataFrame([[row['time'], row['out']]], columns=['time', 'out']), ignore_index=True)
    
# ls
```

### merge time to day

```python
day_flow = {}
for key, value in ls.items():
    tmpDict = {}
    tmp =  pd.DataFrame(dict(values=np.zeros(32), date=pd.date_range('2017-07-01', periods=32, freq='D')))
    
    for index, row in value.iterrows():
        date = row['time'].date()
        out = row['out']
        if date not in tmpDict:
            tmpDict[date] = 0
        tmpDict[date] += out
    for key2, value2 in tmpDict.items():
        tmp.set_value(index=key2.day-1, col='values', value=value2)
    day_flow[key] = tmp
for key, value in day_flow.items():
    print(key)
    print(value)
```

## 切 frame 建 table

試著用三天為 feature，第四天為 target

```python
from collections import deque

for key, value in day_flow.items():
    print(key)
    df = pd.DataFrame(columns=['first', 'second', 'third', 'target'])
    tmp =  deque()
    for index, row in value.iterrows():
        tmp.append(row['values'].item())
        if len(tmp) == 4:
            df = df.append(pd.DataFrame([[tmp[0], tmp[1], tmp[2], tmp[3]]], columns=['first', 'second', 'third', 'target']))
            tmp.popleft()
    print(df)
```
